## Step 1: Explore the Environment

```python
env = JobSchedulingEnv(n_jobs=6, n_machines=3, seed=42)
state = env.reset()
print(env.render())
print(f'\nInitial state: {state}')
print(f'Available actions: {env.get_available_actions()}')
print(f'Lower bound: {env.optimal_lower_bound:.1f}')
```

```python
# Take a step manually
action = 2   # schedule job 2 first
next_state, reward, done, info = env.step(action)
print(f'Scheduled job {action}')
print(f'Reward: {reward}  |  Done: {done}')
print(env.render())
```

## Step 2: Compare Baselines (before training)

```python
env2 = JobSchedulingEnv(n_jobs=6, n_machines=3, seed=42)
env2.reset()
fixed_times = env2.processing_times.copy()
print(f'Processing times: {fixed_times.astype(int).tolist()}')

env2._fixed_times = fixed_times
spt_policy(env2)
print(f'SPT makespan: {env2.get_makespan():.1f}')

env2._fixed_times = fixed_times
lpt_policy(env2)
print(f'LPT makespan: {env2.get_makespan():.1f}')
```
